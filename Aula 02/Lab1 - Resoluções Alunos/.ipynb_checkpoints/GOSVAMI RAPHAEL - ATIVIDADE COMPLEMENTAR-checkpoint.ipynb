```python
titanic3 = pd.read_csv("titanic3.csv")
#Importando dataset
```

```python
#Criando um novo dataset, que recebe o dataset anterior, substituindo os dados faltantes pela media.

titanic_imputted = titanic3.fillna(titanic3.mean()) #imputando a media nos valores faltantes
```

```python
#transformando a coluna boat em numeric
titanic_imputted["boat"] = pd.to_numeric(titanic_imputted['boat'], errors="coerce")
```

```python
#aplicando a mediana nos valores vazios da coluna boat
titanic_imputted = titanic_imputted.fillna(titanic_imputted['boat'].mean())
#titanic_imputted = titanic3.fillna(titanic3.mean())
```

```python
titanic_imputted.head() #lendo os dados novamente
```

```python
#Transformando a class sexo (sex) em dummie (0 ou 1)
titanic_imputted['sex'] = pd.get_dummies(titanic_imputted.sex)
```

```python
#Transformando a class pclass (pclass) em dummie (0 ou 1)
titanic_imputted['pclass'] = pd.get_dummies(titanic_imputted.pclass)
```

```python
#Preditores
preditores = titanic_imputted[['pclass', 'sex', 'boat']]
```

```python
#Alvo
alvo = titanic_imputted['survived']
```

```python
# Divisão dos conjuntos de treino e teste
preditores_treino, preditores_teste, alvo_treino, alvo_teste1 = train_test_split(preditores, alvo, test_size=0.30, random_state=0)
```

```python
# Instancia o modelo de regressão logística com parâmetros default
modelo_reglog = LogisticRegression()

# Treina o modelo com base nos preditores e variável alvo
modelo_reglog.fit(preditores_treino, alvo_treino)

# Prevê, para os dados do conjunto de teste, resultados para a variável alvo
alvo_pred = modelo_reglog.predict(preditores_teste)
```

```python
# Matriz de confusão

print(confusion_matrix(alvo_teste1, alvo_pred))
```

```python
matriz_confusao = pd.crosstab(alvo_teste1, alvo_pred, rownames=['Atual'], colnames=['Previsto'])

# Gerando o gráfico da matriz de confusão
sn.heatmap(matriz_confusao, annot=True)
```

```python
# Acurácia do modelo
print('Acurácia titanic: ', metrics.accuracy_score(alvo_teste1, alvo_pred))
```

```python
print('Precisão: ', metrics.precision_score(alvo_teste1, alvo_pred))

print('Revocação: ', metrics.recall_score(alvo_teste1, alvo_pred))

print('F1-Measure: ', metrics.f1_score(alvo_teste1, alvo_pred))

print('ROC AUC: ', metrics.roc_auc_score(alvo_teste1, alvo_pred))
```

```python
# Separando a variável alvo das variáveis independentes
x = titanic_f.drop(["survived"],axis=1)
y = titanic_f["survived"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.40, random_state=0)
```

```python
def confusion_matrix_model(model_used):
    cm = confusion_matrix(y_test, model_used.predict(x_test))
    col = ["Previsto Morto","Previsto Sobrevivente"]
    cm = pd.DataFrame(cm)
    cm.columns = ["Previsto Morto","Previsto Sobrevivente"]
    cm.index=["Realmente Morto","Realmente Sobrevivente"]
    cm[col]=np.around(cm[col].div(cm[col].sum(axis=1),axis=0),decimals=2)
    return cm
```

```python
log_reg = LogisticRegression()

log_reg = log_reg.fit(x_train, y_train)

log_reg
```

```python
# Coeficientes relativos as variáveis independentes/preditoras
print('Coeficientes: \n', log_reg.coef_)


print('Interfacept: \n', log_reg.intercept_)
```

```python
# Prevê, para os dados do conjunto de teste, resultados para a variável alvo
predicao_reglog = log_reg.predict(x_test)
```

```python
def plot_roc_curves(fper1, tper1, title1, fper2, tper2, title2, fper3, tper3, title3):
    plt.plot(fper1, tper1, color="orange", label=title1)
    plt.plot(fper2, tper2, color="green", label=title2)
    plt.plot(fper3, tper3, color="red", label=title3)
    plt.plot([0,1], [0,1], color='darkblue', linestyle='--')
    plt.xlabel('Taxa de Falsos Positivos')
    plt.ylabel('Taxa de Falsos Negativos')
    plt.title("Curva ROC")
    plt.legend()
    plt.show()
```
