##### 4 - Interpretação do sumário.

```python
# Removi as classes que não representam nenhuma relação com sobrevivência, realizei conversão para variáveis dummies e 
# inseri os dados tratados em um novo dataFrame.

data_class = data_titanic.survived
dados_preditoras = data_titanic.drop(['survived','name','home.dest','embarked','ticket','cabin','parch','boat','body'], axis=1)

dados_preditoras['sex'] = pd.get_dummies(dados_preditoras['sex'])
#dados_preditoras['boat'] = pd.get_dummies(dados_preditoras['boat'])
#dados_preditoras['cabin'] = pd.get_dummies(dados_preditoras['cabin'])

data_titanic_tratado = dados_preditoras

data_titanic_tratado['survived'] = data_class

dados_preditoras.drop(['survived'], axis=1, inplace=True)

# Retorna dados que representam centralidade estatística como média, mediana e desvio padrão além dos valores máximos e mínimos
# e porcentagem representativa de acumulo de dados (25%, 50%, 75%) além do total de registros.

dados_preditoras.fillna(dados_preditoras.mean(), inplace=True)

#data_titanic_tratado.describe()

dados_preditoras
```

##### 5 - Tenho relação/correlação forte entre variáveis do meu dataset ?

```python
# Analisando a correlação entre as variáveis através do método de pearson.

# Esta função serve para comparar a relação/correlação entre duas variaáeis, basicamente ela divide os desvios padrões das
# variáveis, retornando o valor númerico que poderá variar entre -1 a 1, onde -1  simboliza anticorrelação perfeita, 1
# correlação perfeita e 0 significa que essas variáveis não possuem relação linear, porém, podem possuir outro tipo de 
# dependência que seja não linear, assim sendo, um estudo mais aprofundado deve ser feito.

data_titanic_tratado.corr(method='pearson')
```

##### 6 - Quais as variáveis imagino que terão influencia na pergunta que desejo responder ?

```python
data_titanic_tratado.fillna(0).corr().round(2)
matriz_correlacao = data_titanic_tratado.corr().round(2)

fig, ax = plt.subplots(figsize=(16,10)) 
sb.heatmap(data = matriz_correlacao,  linewidths=.5, annot = True)
```

```python
preditores_treino, preditores_teste, alvo_treino, alvo_teste = train_test_split(dados_preditoras, data_class, test_size=0.30, random_state=0)

# Instancia o modelo de regressão logística com parâmetros default
modelo_reglog = LogisticRegression()

# Treina o modelo com base nos preditores e variável alvo
modelo_reglog.fit(preditores_treino, alvo_treino)

# Prevê, para os dados do conjunto de teste, resultados para a variável alvo
alvo_pred = modelo_reglog.predict(preditores_teste)

#confusion_matrix(alvo_teste, alvo_pred)

print('Precisão: ', metrics.precision_score(alvo_teste, alvo_pred))

print('Revocação: ', metrics.recall_score(alvo_teste, alvo_pred))

print('F1-Measure: ', metrics.f1_score(alvo_teste, alvo_pred))

print('ROC AUC: ', metrics.roc_auc_score(alvo_teste, alvo_pred))

data_teste = preditores_teste
data_teste['survived'] = alvo_teste
data_teste['Previsto'] = alvo_pred
data_teste
```
